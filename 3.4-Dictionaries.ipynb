```python
#accessing dictionary elements 

student={"name": "Pratham" , "age" : 21 , "grade":'A'} 

print(student['name'])
print(student['age'])
print(student['grade'])
```

```python
print(student.get('grade'))
print(student.get('name'))
print(student.get('age'))
```

```python
#Modifying dictionary elements

student['age']= 25 # update value for the key 

print(student)

```

```python
student['address']='India' # add a new key and value

print(student) 
```

```python
del student['address'] # delete the key value pair

print(student)

```

```python
keys = student.keys()
print(keys)

values = student.values()
print(values)

items = student.items()
print(items)
```

```python

student_copy=student

print(student)
print(student_copy)
```

```python
student['age']=23
print(student)
print(student_copy)
```

```python
new_student_copy = student.copy()
student['grade']='AA'
print(student)
print(new_student_copy)
```

```python
for keys in student.keys():
    print(keys)

for values in student.values():
    print(values)    

for keys, values in student.items():
    print(f"{keys}:{values}")    
```

```python
students ={
    'student1':{'name':'Pratham','age': 21},
    'student2':{'name':'Peter','age': 22}
}
print(students)
```

```python
print(students['student1']['name'])
print(students['student1']['age'])
```

```python
for student_id , student_info in students.items():
    print(f"{student_id} : {student_info}")
    for key , value in student_info.items():
        print(f"{key}:{value}")
```

```python
# dictionary comprehension

squares={x:x**2 for x in range (5)}
print(squares)
```

```python
even_squares={x:x**2 for x in range (10) if x%2==0}
print(even_squares)
```

```python
numbers=[1,1,2,2,2,3,3,4,4,4,3,4,4,2,4,4,7]
frequency ={}

for num in numbers:
    if num in frequency:
        frequency[num]+=1
    else:
        frequency[num]=1
        
        
        
        
        
    
    
print(frequency)    
```
