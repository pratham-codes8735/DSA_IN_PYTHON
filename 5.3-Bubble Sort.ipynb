```python
def bubble_sort(arr):
    n=len(arr)
    for i in range(0, n):
        for j in range (0 , n-1-i):
            if (arr[j]>arr[j+1]):
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr            


unsorted_list=[266,56,87,49,12,54,87]
sorted_list=bubble_sort(unsorted_list)
print("Sorted_list : " , sorted_list)
```
